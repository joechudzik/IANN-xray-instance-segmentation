```python
import pandas as pd
import os
```

```python
path = 'C:\\Users\\dental-1\\Documents\\dent_seg\\1_helper_code\\labels\\'
bounding_boxes = os.listdir(path)
```

```python
bounding_boxes
```

```python
for bb in bounding_boxes:
    if 'txt' in bb:
        # read in the bounding box data
        df = pd.read_csv(path + bb, header=None, sep=' ', dtype=float)
        # if there's the sixth confidence column then drop it
        if len(df.columns) == 6:
            df.drop(columns=[5], inplace=True)
            df.to_csv(path + bb, index=False, header=False, sep=' ')
```

```python
for bb in bounding_boxes:
    if 'txt' in bb:
        print(bb)
        df = pd.read_csv(bb, header=0, sep=',')
        if df.columns[0] == 'Unnamed: 0':
            df.drop(columns=['Unnamed: 0'], inplace=True)
            df['0'] = int(0)
            df.to_csv(bb, index=False, header=False, sep=' ')
```
