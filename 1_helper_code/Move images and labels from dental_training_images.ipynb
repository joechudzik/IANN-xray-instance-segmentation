```python
# get name of dental training image files
files = glob("D:/dental_training_images/*")
```

```python
# move raw images
for file in files:
    if 'panoramic' in file and file[-1:] in ['0','1','2','3','4','5','6','7','8']:
        raw = glob(file+'/*')
        destination = 'C:\\Users\\dental-1\\Documents\\dent_seg\\dental images\\raw_images'
        for image in raw:
            source = image
            shutil.copy(source, destination)
        print('Done moving for folder {}'.format(file))
```

```python
# move clahe images
for file in files:
    if 'panoramic' in file and 'clahe' in file:
        raw = glob(file+'/*')
        destination = 'C:\\Users\\dental-1\\Documents\\dent_seg\\dental images\\clahe_images'
        for image in raw:
            source = image
            shutil.copy(source, destination)
        print('Done moving for folder {}'.format(file))
```

```python
# move label files
for file in files:
    if 'panoramic' in file and 'labels' in file:
        print(file)
        label_files = glob(file + "/*")
        for label in label_files:
            if 'classes' not in label and 'xml' not in label:
                label_df = pd.read_csv(label, 
                                   header=None, 
                                   sep=' ', 
                                   dtype=float, 
                                   names=['label_num','x_center','y_center','width','height'])
                label_df['label_num'] = 0
                label_name = label.split('\\')[-1]
                # save bounding box file to new folder
                filename = 'C:\\Users\\dental-1\\Documents\\dent_seg\\train_yolo_tooth_segmentation\\bounding_boxes\\' + label_name
                label_df.to_csv(filename, index=False, header=False, sep=' ')
        print('Done moving for folder {}'.format(file))
```
