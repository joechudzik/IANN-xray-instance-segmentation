```python
import pandas as pd
import os
import numpy as np
import cv2
```

```python
bounding_boxes = os.listdir('C:\\Users\\dental-1\\Documents\\dent_seg\\1_helper_code\\labels\\')
bounding_boxes
```

```python
for bb in bounding_boxes:
    if 'txt' in bb:
        filename = 'C:\\Users\\dental-1\\Documents\\dent_seg\\1_helper_code\\labels\\' + bb
        # get df
        df = pd.read_csv(filename, sep=' ', header=None)
        df.rename(columns={0: 'class', 1: 'x_center', 2: 'y_center', 3: 'height', 4: 'width'}, inplace=True)
        # get top/bottom
        average_y = df['y_center'].mean()
        df['top/bottom'] = np.where(df.y_center < average_y, 'top', 'bottom')
        # sort teeth
        df.sort_values(by=['top/bottom', 'x_center'], inplace=True)
        # drop top/bottom column
        df.drop(columns=['top/bottom'], inplace=True)
        df.to_csv(filename, index=False, header=False, sep=' ')
```
