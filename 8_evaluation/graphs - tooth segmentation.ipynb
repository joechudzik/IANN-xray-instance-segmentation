```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\dental-1\\Documents\\dent_seg\\5_train_yolo_tooth_segmentation\\yolov5\\runs\\train\\teeth_detect4\\results.csv')
df.head()
```

```python
plt.figure(figsize=(10,7))
ax = sns.lineplot(data=df[['      metrics/recall']], palette='Greens', legend=False)
ax.set_title('Recall of Test Data', fontsize = 20)
ax.set_xlabel('Epoch', fontsize=15)
ax.set_ylabel('Results', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,7))
ax = sns.lineplot(data=df[['   metrics/precision']], palette='Greens', legend=False)
ax.set_title('Precision of Test Data', fontsize = 20)
ax.set_xlabel('Epoch', fontsize=15)
ax.set_ylabel('Results', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,7))
ax = sns.lineplot(data=df[['   metrics/precision', '      metrics/recall']], palette='Paired_r', legend=False)
ax.set_title('Precision Recall of Test Data', fontsize = 20)
ax.set_xlabel('Epoch', fontsize=15)
ax.set_ylabel('Results', fontsize=15)
plt.legend(loc='lower right', labels=['Precision', 'Recall'], fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,7))
ax = sns.lineplot(data=df[['     metrics/mAP_0.5']], palette='Greens', legend=False)
ax.set_title('mAP 0.5 of Test Data', fontsize = 20)
ax.set_xlabel('Epoch', fontsize=15)
ax.set_ylabel('Results', fontsize=15)
plt.show()
```
